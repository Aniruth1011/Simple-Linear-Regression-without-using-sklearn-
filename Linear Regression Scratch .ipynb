```python
import pandas as pd
import numpy as np
#import seaborn as sns
import matplotlib.pyplot as plt
import random
from sklearn.model_selection import train_test_split
```

```python
d = pd.read_csv(r"D:\Downloads\regression (2).csv")
```

```python
x = d['YearsExperience']
y = d['Salary']
```

```python
plt.scatter(x , y)
plt.title('Ground Truth')
```

We Know that for a simple linear regression model , we can find the line of best fit by calculating m and c

Σ(y) = mΣx + nc

Σ(xy) = mΣ(x2) + cΣx

```python
sum_y = sum(y)
sum_x = sum(x)
sum_xy = sum(x*y)
sum_x2 = sum(x**2)
n = len(x)
```

```python
sum_y
```

```python
sum_x
```

```python
sum_xy
```

```python
sum_x2
```

```python
n
```

```python
A = np.array([[sum_x , n] , [sum_x2 , sum_x]])
```

```python
B = np.array([sum_y , sum_xy])
```

```python
output = np.linalg.solve(A, B)
```

```python
output
```

```python
m_best = output[0]
c_best = output[1]
```

```python
y_pred = m_best * x + c_best
```

```python
plt.scatter(x , y)
plt.plot(x, y_pred , c = 'r')
plt.title('Line of Best Fit')
```

```python
SSE = sum((y - y_pred)/y)*100/n
```

```python
print("Mean % Error is : " , SSE)
```

This shows that our line of best fit is really good .
